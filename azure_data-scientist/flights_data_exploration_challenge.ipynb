```
#replace null with mean values
df_flights['DepDel15'] = df_flights['DepDel15'].fillna(df_flights['DepDel15'].mean())


# identify outliers in DepDelay & ArrDelay
# upper_depdelay = df_flights.DepDelay.quantile(0.99)
# lower_depdelay = df_flights.DepDelay.quantile(0.01)

# upper_arrdelay = df_flights.ArrDelay.quantile(0.99)
# lower_arrdelay = df_flights.ArrDelay.quantile(0.01)

def get_count_from_quantile(df, column_name):
    df_column = df[column_name]
    upper = df_column.quantile(0.99)
    lower = df_column.quantile(0.01)
    num_outliers = df[df_column > upper][column_name].count() + df[lower > df_column][column_name].count()
    return num_outliers

def filter_from_quantile(df, column_names, get_count=False):
    for column_name in column_names:
        df_column = df[column_name]
        upper = df_column.quantile(0.99)
        lower = df_column.quantile(0.01)
        df = df[(df_column < upper) & (df_column > lower)]
    if get_count:
        return df.count()
    return df




num_outliers = get_count_from_quantile(df_flights, "DepDelay")
print(num_outliers)
```

```
df_flights_refined = filter_from_quantile(df_flights, ["DepDelay", "ArrDelay"])

df_flights_refined.head()
print("Dropped", str(df_flights["DepDelay"].count() - df_flights_refined["DepDelay"].count()), " many rows. Out of", str(df_flights["DepDelay"].count()))
```

```
def show_distribution(column):
    """
    This function will make a distribution graph and display it
    """
    
    #Get the min, max, mean, median, and mode values
    min_val = float(column.min())
    max_val = float(column.max())
    mean_val = float(column.mean())
    mode_val = float(column.mode())
    median_val = float(column.median())
    
    
    #print them
    print("Min: {:.2f}\n Max: {:.2f}\n Mean: {:.2f}\n Mode: {:.2f}\n Median: {:.2f}\n".format(
              min_val,
              max_val,
              mean_val,
              mode_val,
              median_val))
    
    # Create a figure for 2 subplots (2 rows, 1 column)
    fig, ax = plt.subplots(1,2, figsize=(17,6))
    
    
    # Plot the histogram, to ax[0]
    ax[0].hist(column)
    
    # Add lines for mean, median, and mode
    for val in [min_val, max_val, mean_val, mode_val, median_val]:
        ax[0].axvline(val, linestyle="dashed", color=np.random.rand(3,))
    
    # Plot a boxplot
    ax[1].boxplot(column, vert=False)
    
    #Add a title
    fig.suptitle("Data distribution")
    
    #show the figure
    fig.show()
```

```
#C0MPARISON OF REFINING VS NOT REFINING
show_distribution(df_flights["DepDelay"])
show_distribution(df_flights_refined["DepDelay"])
show_distribution(df_flights_refined["ArrDelay"])
```

```
def summary_by_column(df, column_name, to_find="ArrDelay"):
    entries = list(df[column_name].unique())
    return_dict = {}
    for ent in entries:
        return_dict[ent] = round(df[df[column_name]==ent][to_find].mean(),2)
    return return_dict
```
